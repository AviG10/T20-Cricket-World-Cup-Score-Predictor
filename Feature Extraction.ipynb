```python
import pandas as pd
import pickle
import numpy as np
```

```python
df = pickle.load(open('dataset_level2.pkl','rb'))
```

```python
df

# batting_team
# bowling team
# city
# current score
# ball left
# wickets left
# current rr
# last five
```

```python
df.isnull().sum()
```

```python
#venue at which city is missing
df[df['city'].isnull()]['venue'].value_counts()
```

```python
#replace city by first word of venue
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
df['city'] = cities
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['venue'],inplace=True)
```

```python
df
```

```python
#we will consider those cities where minimum 5 matches happens that is minimum 600 balls bowled
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df
```

```python
#extracting current score by cummulative sum
df['current_score'] = df.groupby('match_id').cumsum()['runs']
```

```python
df
```

```python
#split ball coloumn by dot(.) where left part will be over and right part will be balls
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
# if balls left is less than zero then i am keeping it as zero
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
#if player dismissed is zero then keep it as zero otherwise replace it with 1 and 
# then find cummulative sum and subtract from 10 to get wickets left
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id').cumsum()['player_dismissed']
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
#creating current run rate coloumn
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
# calculating last five over or 30 balls sum . for firt 30 balls there will be nolast five over sum
#so it will be Nan for first 30 balls
groups = df.groupby('match_id') #group data on the basis of match id

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
```

```python
df['last_five'] = last_five
```

```python
#df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','']]
```

```python
# calculate final score in every match
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
#drop all missing values
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
#shuffling the values so that no baised developed
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df
```

```python
X = final_df.drop(columns=['runs_x']) #extracting x
y = final_df['runs_x'] #extracting y 
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
# from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
#As 'batting_team','bowling_team','city'these column is in string so we have to oneHotEncode it that is
# encode categorical features as one-hot numeric array and passthrough rest columns

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
#creating pipeline
pipe = Pipeline(steps=[
    ('step1',trf), #oneHotEncode first
    ('step2',StandardScaler()), #Scaling the data
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1)) #run XGboost Regressor next
])

# n_estimators = no. of trees
# learning_rate = shrinkage we do at every step we are making
# max_depth = maximum depth of the tree i.e the longest path between root and leaf node
#  random_state = provides seed value to function's internal random number generator
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))

# ----------------------r2_score---------------------
#r2_score is the amount of the variation in the output dependent attribute which is predictable
# from the input independent variable(s).
#R2= 1- SSres / SStot ,SSres is the sum of squares of the residual errors and 
# SStot is the total sum of the errors.
#It is used to check how well-observed results are reproduced by the model, 
# depending on the ratio of total deviation of results described by the model.


# ----------------------mean absolute error---------------------
#Mean Absolute Error calculates the average difference between the calculated values and actual values. 
#It calculates errors between actual values and values predicted by the model. 
#Mean Absolute Error = (1/n) * Σ|yi – xi|
# Σ: Greek symbol for summation
# yi: Actual value for the ith observation
# xi: Calculated value for the ith observation
# n: Total number of observations
```

```python
pickle.dump(pipe,open('pipe.pkl','wb'))
```
